```python
def djikstra(graph, start, end):
    n = len(graph)
    visited = [False] * n
    distance = [float('inf')] * n
    previous = [None] * n
    distance[start] = 0

    for _ in range(n):
        min_distance = float('inf')
        u = -1
        
        for v in range(n):
            if not visited[v] and distance[v] < min_distance:
                min_distance = distance[v]
                u = v
        
        if u == -1:  # Jika tidak ada node yang tersisa
            break

        visited[u] = True
        
        for v in range(n):
            if graph[u][v] > 0 and not visited[v]:
                new_distance = distance[u] + graph[u][v]
                if new_distance < distance[v]:
                    distance[v] = new_distance
                    previous[v] = u
    
    # Membangun rute dari start ke end
    path = []
    current = end
    while current is not None:
        path.append(current)
        current = previous[current]
    path.reverse()  # Membalikkan urutan path

    return distance, path
```

```python
cities = ["Surabaya", "Malang", "Semarang", "Jakarta", "Yogyakarta", "Bandung"]
n = len(cities)

# Membuat graf dengan bobot
graph = np.zeros((n, n))

# Misalnya kita masukkan bobot seperti berikut (dalam km)
# Memperbarui edges untuk memberikan rute lebih kompleks
edges = [
    (0, 1, 90),   # Surabaya ke Malang
    (0, 2, 320),  # Surabaya ke Semarang
    (1, 2, 230),  # Malang ke Semarang
    (1, 3, 300),  # Malang ke Jakarta (langsung)
    (2, 3, 480),  # Semarang ke Jakarta
    (3, 4, 570),  # Jakarta ke Yogyakarta
    (2, 4, 400),  # Semarang ke Yogyakarta
    (4, 5, 150),  # Yogyakarta ke Bandung
    (3, 5, 150)   # Jakarta ke Bandung
]


for edge in edges:
    graph[edge[0]][edge[1]] = edge[2]
    graph[edge[1]][edge[0]] = edge[2]  # Jika grafnya tidak terarah
```

```python
start_city = input("Masukkan kota awal (Surabaya, Malang, Semarang, Jakarta, Yogyakarta, Bandung):")
end_city = input("Masukkan kota akhir (Surabaya, Malang, Semarang, Jakarta, Yogyakarta, Bandung):")

#menggunakan index untuk mendapatkan index kota dalam daftar
start_node = cities.index(start_city) 
end_node = cities.index(end_city)

distances, path = djikstra(graph, start_node, end_node)

print(f"Jarak terpendek dari {start_city} ke {end_city}: {distances[end_node]} km")
print("Rute terpendek:", " -> ".join(cities[i] for i in path))

# Menampilkan setiap kota yang dilalui
print("Kota-kota yang dilalui dalam rute terpendek:")
for i in path:
    print(f"- {cities[i]}")

# Visualisasi graf
G = nx.Graph()

# Menambahkan node
for city in cities:
    G.add_node(city)

# Menambahkan edge
for edge in edges:
    G.add_edge(cities[edge[0]], cities[edge[1]], weight=edge[2])

# Mengatur posisi node
pos = nx.spring_layout(G)

plt.figure(figsize=(10, 8))
nx.draw(G, pos, with_labels=True, node_size=700, node_color='lightblue', font_size=12, font_weight='bold')
labels = nx.get_edge_attributes(G, 'weight')
nx.draw_networkx_edge_labels(G, pos, edge_labels=labels, font_color='red')
plt.title('Graf Jalur Kota di Indonesia dengan Algoritma Dijkstra')
plt.show()
```
